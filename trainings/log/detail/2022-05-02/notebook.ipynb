# ``while`` loops, and ``else``

```python
import random

ntries = 0
win = False
while ntries <= 6:
    ntries += 1
    eyes = random.randrange(1,7)
    if eyes == 6:
        win = True
        break
if win:
    print('hooray')
else:
    print('lose')
```

# ``for`` loops (and ``else``)

```python
import random

for _ in range(6):
    eyes = random.randrange(1,7)
    if eyes == 6:
        print('hooray')
        break
else:
    print('lose')
```

# ``yield``

```python
def even(start, end):
    print('even: entered')
    run = start
    while run < end:
        if run % 2 == 0:
            print('even: yielding')
            yield run
            print('even: continuing after yield')
        run += 1
    print('even: terminating')
```

```python
for element in even(2, 10):
    print(element)
```

```python
e = even(2, 10)
type(e)
```

```python
it = iter(e)
```
